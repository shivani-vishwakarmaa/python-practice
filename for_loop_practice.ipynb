# 游대 For Loop Practice in Python

This notebook contains a collection of examples using `for` loops in Python.
Each code cell demonstrates a specific use case such as:

- Iterating over lists and strings
- Mathematical operations
- Pattern printing
- Fibonacci, factorial, power, GCD, Armstrong number

Use this for hands-on `for` loop mastery!


```python
# 游대 Print list elements using for loop
l = [1, 4, 6, 8, 9]
for item in l:
    print(item)
```

```python
# 游댌 Find a number in list using for loop
l = [1, 4, 6, 8, 9]
v = 40
for item in l:
    if item == v:
        print("found")
        break
else:
    print("not found")
```

```python
# 游대 Practice: Sum of first N numbers
n = int(input("Enter a number: "))
total = 0
for i in range(1, n + 1):
    total += i
print("Sum is:", total)
```

```python
# 游대 Practice: Guess the secret number (fixed number of chances)
secret = 7
for _ in range(5):
    guess = int(input("Guess the number: "))
    if guess == secret:
        print("You guessed it right!")
        break
    else:
        print("Wrong! Try again.")
```

```python
# 游대 Factorial using for loop
n = 5
fact = 1
for i in range(1, n + 1):
    fact *= i
print("Factorial:", fact)
```

```python
# 游대 Even numbers from 1 to 10 using for loop
for i in range(1, 11):
    if i % 2 == 0:
        print(i, end=" ")
```

```python
# 游대 Count digits in a number using for loop
num = 123456
count = 0
for _ in str(num):
    count += 1
print("\nTotal digits:", count)
```

```python
# 游대 Palindrome check using for loop
num = 121
rev = ""
for ch in str(num):
    rev = ch + rev
print("Palindrome:", str(num) == rev)
```

```python
# 游대 Multiplication table of a number using for loop
n = 7
for i in range(1, 11):
    print(f"{n} x {i} = {n*i}")
```

```python
# 游대 GCD using Euclidean Algorithm (while preferred, for shown for syntax)
a, b = 36, 60
for _ in range(100):  # limit iterations
    if b == 0:
        break
    a, b = b, a % b
print("GCD is:", a)
```

```python
# 游대 Fibonacci series up to N terms using for loop
n = 10
a, b = 0, 1
for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
# 游대 Power using for loop (a^b)
a = 2
b = 5
result = 1
for _ in range(b):
    result *= a
print("Power:", result)
```

```python
# 游대 Sum of digits of a number using for loop
num = 9876
total = 0
for ch in str(num):
    total += int(ch)
print("Sum of digits:", total)
```

```python
# 游대 Print pattern using for loop
rows = 3
for i in range(1, rows + 1):
    print("*" * i)
```

```python
# 游대 Armstrong number (3-digit)
num = 153
sum_of_cubes = 0
for digit in str(num):
    sum_of_cubes += int(digit) ** 3
print("Armstrong:", num == sum_of_cubes)
```
