# ðŸ” For Loop Practice in Python

This notebook contains a collection of examples using `for` loops in Python.
Each code cell demonstrates a specific use case such as:

- Iterating over lists and strings
- Mathematical operations
- Pattern printing
- Fibonacci, factorial, power, GCD, Armstrong number

Use this for hands-on `for` loop mastery!


```python
# ðŸ” Print list elements using for loop
l = [1, 4, 6, 8, 9]
for item in l:
    print(item)
```

```python
# ðŸ” Find a number in list using for loop
l = [1, 4, 6, 8, 9]
v = 40
for item in l:
    if item == v:
        print("found")
        break
else:
    print("not found")
```

```python
# ðŸ” Practice: Sum of first N numbers
n = int(input("Enter a number: "))
total = 0
for i in range(1, n + 1):
    total += i
print("Sum is:", total)
```

```python
# ðŸ” Practice: Guess the secret number (fixed number of chances)
secret = 7
for _ in range(5):
    guess = int(input("Guess the number: "))
    if guess == secret:
        print("You guessed it right!")
        break
    else:
        print("Wrong! Try again.")
```

```python
# ðŸ” Factorial using for loop
n = 5
fact = 1
for i in range(1, n + 1):
    fact *= i
print("Factorial:", fact)
```

```python
# ðŸ” Even numbers from 1 to 10 using for loop
for i in range(1, 11):
    if i % 2 == 0:
        print(i, end=" ")
```

```python
# ðŸ” Count digits in a number using for loop
num = 123456
count = 0
for _ in str(num):
    count += 1
print("\nTotal digits:", count)
```

```python
# ðŸ” Palindrome check using for loop
num = 121
rev = ""
for ch in str(num):
    rev = ch + rev
print("Palindrome:", str(num) == rev)
```

```python
# ðŸ” Multiplication table of a number using for loop
n = 7
for i in range(1, 11):
    print(f"{n} x {i} = {n*i}")
```

```python
# ðŸ” GCD using Euclidean Algorithm (while preferred, for shown for syntax)
a, b = 36, 60
for _ in range(100):  # limit iterations
    if b == 0:
        break
    a, b = b, a % b
print("GCD is:", a)
```

```python
# ðŸ” Fibonacci series up to N terms using for loop
n = 10
a, b = 0, 1
for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
# ðŸ” Power using for loop (a^b)
a = 2
b = 5
result = 1
for _ in range(b):
    result *= a
print("Power:", result)
```

```python
# ðŸ” Sum of digits of a number using for loop
num = 9876
total = 0
for ch in str(num):
    total += int(ch)
print("Sum of digits:", total)
```

```python
# ðŸ” Print pattern using for loop
rows = 3
for i in range(1, rows + 1):
    print("*" * i)
```

```python
# ðŸ” Armstrong number (3-digit)
num = 153
sum_of_cubes = 0
for digit in str(num):
    sum_of_cubes += int(digit) ** 3
print("Armstrong:", num == sum_of_cubes)
```

```python
# =====================================================
# ðŸ” REVERSE STRING â€” BACKEND-QUALITY PYTHON SHEET
# =====================================================

from typing import Optional


def reverse_string_basic(s: str) -> str:
    """
    BEST (Pythonic)
    - Uses slicing
    - Clean and fast
    - O(n) time, O(n) space
    """
    return s[::-1]


def reverse_string_safe(s: Optional[str]) -> Optional[str]:
    """
    BACKEND SAFE VERSION
    - Handles None
    - Handles empty string
    - Predictable return type
    """
    if s is None:
        return None
    if not isinstance(s, str):
        raise TypeError("Input must be a string")
    return s[::-1]


def reverse_string_no_slicing(s: str) -> str:
    """
    INTERVIEW VARIANT (No slicing allowed)
    - Avoids O(nÂ²) string concatenation
    - Uses list + join
    - O(n) time, O(n) space
    """
    chars = list(s)
    chars.reverse()
    return "".join(chars)


def reverse_string_with_reversed(s: str) -> str:
    """
    STANDARD LIBRARY APPROACH
    - Uses reversed() + join
    - Clean and readable
    """
    return "".join(reversed(s))


def reverse_string_worst(s: str) -> str:
    """
    WORST PRACTICE (Avoid in backend)
    - O(nÂ²) due to string immutability
    - Shown only for interview discussion
    """
    rev = ""
    for ch in s:
        rev = ch + rev
    return rev


# =====================================================
# âœ… EXAMPLES
# =====================================================
if __name__ == "__main__":
    text = "hello"

    print("Basic (BEST):", reverse_string_basic(text))
    print("Safe (BACKEND):", reverse_string_safe(text))
    print("No slicing:", reverse_string_no_slicing(text))
    print("Using reversed():", reverse_string_with_reversed(text))
    print("Worst (avoid):", reverse_string_worst(text))
```

```python
# =====================================================
# ðŸ” REVERSE LIST â€” BACKEND-QUALITY PYTHON SHEET
# =====================================================

from typing import Iterable, List, Optional, TypeVar

T = TypeVar("T")


def reverse_list_in_place(items: List[T]) -> List[T]:
    """
    BEST (In-place)
    - Uses list.reverse()
    - O(n) time, O(1) space
    - Mutates input (callers must know this)
    """
    if items is None:
        raise ValueError("Input list cannot be None")
    items.reverse()
    return items


def reverse_list_safe(items: Optional[Iterable[T]]) -> Optional[List[T]]:
    """
    BACKEND SAFE VERSION â­
    - Handles None
    - Does NOT mutate input
    - Predictable return type
    """
    if items is None:
        return None
    return list(reversed(items))


def reverse_list_slicing(items: List[T]) -> List[T]:
    """
    PYTHONIC & CLEAN
    - Uses slicing
    - O(n) time, O(n) space
    - Does NOT mutate input
    """
    return items[::-1]


def reverse_list_two_pointer(items: List[T]) -> List[T]:
    """
    DSA / FAANG STYLE
    - In-place reversal using two pointers
    - O(n) time, O(1) space
    """
    if items is None:
        raise ValueError("Input list cannot be None")

    left, right = 0, len(items) - 1
    while left < right:
        items[left], items[right] = items[right], items[left]
        left += 1
        right -= 1

    return items


def reverse_list_loop(items: List[T]) -> List[T]:
    """
    AVERAGE (Loop-based)
    - Uses index loop + append
    - O(n) time, O(n) space
    """
    result: List[T] = []
    for i in range(len(items) - 1, -1, -1):
        result.append(items[i])
    return result


def reverse_list_worst(items: List[T]) -> List[T]:
    """
    WORST PRACTICE ðŸš«
    - Uses insert(0, x) or pop(0)
    - O(nÂ²) time due to shifting
    """
    result: List[T] = []
    for item in items:
        result.insert(0, item)
    return result


# =====================================================
# âœ… EXAMPLES
# =====================================================
if __name__ == "__main__":
    data = [1, 2, 3, 4, 5]

    print("Safe (BEST):", reverse_list_safe(data))
    print("Slicing:", reverse_list_slicing(data))
    print("In-place:", reverse_list_in_place(data.copy()))
    print("Two-pointer:", reverse_list_two_pointer(data.copy()))
    print("Loop:", reverse_list_loop(data))
    print("Worst (avoid):", reverse_list_worst(data))
```
