# üîÅ For Loop Practice in Python

This notebook contains a collection of examples using `for` loops in Python.
Each code cell demonstrates a specific use case such as:

- Iterating over lists and strings
- Mathematical operations
- Pattern printing
- Fibonacci, factorial, power, GCD, Armstrong number

Use this for hands-on `for` loop mastery!


```python
# üîÅ Print list elements using for loop
l = [1, 4, 6, 8, 9]
for item in l:
    print(item)
```

```python
# üîç Find a number in list using for loop
l = [1, 4, 6, 8, 9]
v = 40
for item in l:
    if item == v:
        print("found")
        break
else:
    print("not found")
```

```python
# üîÅ Practice: Sum of first N numbers
n = int(input("Enter a number: "))
total = 0
for i in range(1, n + 1):
    total += i
print("Sum is:", total)
```

```python
# üîÅ Practice: Guess the secret number (fixed number of chances)
secret = 7
for _ in range(5):
    guess = int(input("Guess the number: "))
    if guess == secret:
        print("You guessed it right!")
        break
    else:
        print("Wrong! Try again.")
```

```python
# üîÅ Factorial using for loop
n = 5
fact = 1
for i in range(1, n + 1):
    fact *= i
print("Factorial:", fact)
```

```python
# üîÅ Even numbers from 1 to 10 using for loop
for i in range(1, 11):
    if i % 2 == 0:
        print(i, end=" ")
```

```python
# üîÅ Count digits in a number using for loop
num = 123456
count = 0
for _ in str(num):
    count += 1
print("\nTotal digits:", count)
```

```python
# üîÅ Palindrome check using for loop
num = 121
rev = ""
for ch in str(num):
    rev = ch + rev
print("Palindrome:", str(num) == rev)
```

```python
# üîÅ Multiplication table of a number using for loop
n = 7
for i in range(1, 11):
    print(f"{n} x {i} = {n*i}")
```

```python
# üîÅ GCD using Euclidean Algorithm (while preferred, for shown for syntax)
a, b = 36, 60
for _ in range(100):  # limit iterations
    if b == 0:
        break
    a, b = b, a % b
print("GCD is:", a)
```

```python
# üîÅ Fibonacci series up to N terms using for loop
n = 10
a, b = 0, 1
for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
# üîÅ Power using for loop (a^b)
a = 2
b = 5
result = 1
for _ in range(b):
    result *= a
print("Power:", result)
```

```python
# üîÅ Sum of digits of a number using for loop
num = 9876
total = 0
for ch in str(num):
    total += int(ch)
print("Sum of digits:", total)
```

```python
# üîÅ Print pattern using for loop
rows = 3
for i in range(1, rows + 1):
    print("*" * i)
```

```python
# üîÅ Armstrong number (3-digit)
num = 153
sum_of_cubes = 0
for digit in str(num):
    sum_of_cubes += int(digit) ** 3
print("Armstrong:", num == sum_of_cubes)
```

```python
# =====================================================
# üîÅ STRING REVERSAL ‚Äî ALL METHODS (INTERVIEW READY)
# =====================================================

s = "hello"

# -----------------------------------------------------
# ‚úÖ METHOD 1: SLICING  ‚≠ê BEST (Pythonic & Fastest)
# Time: O(n), Space: O(n)
# -----------------------------------------------------
rev1 = s[::-1]
print("Method 1 (BEST):", rev1)


# -----------------------------------------------------
# ‚úÖ METHOD 2: reversed() + join  ‚≠ê BEST (No slicing)
# Time: O(n), Space: O(n)
# -----------------------------------------------------
rev2 = "".join(reversed(s))
print("Method 2 (BEST):", rev2)


# -----------------------------------------------------
# ‚úÖ METHOD 3: LOOP + LIST + join  ‚≠ê EFFICIENT (No O(n¬≤))
# Time: O(n), Space: O(n)
# -----------------------------------------------------
chars = []
for ch in s:
    chars.append(ch)

chars.reverse()
rev3 = "".join(chars)
print("Method 3 (GOOD):", rev3)


# -----------------------------------------------------
# ‚ö†Ô∏è METHOD 4: INDEX LOOP + STRING CONCAT  ‚ùå AVERAGE
# Time: O(n¬≤), Space: O(n)
# Reason: String immutability
# -----------------------------------------------------
rev4 = ""
for i in range(len(s)):
    rev4 += s[len(s) - 1 - i]

print("Method 4 (AVERAGE):", rev4)


# -----------------------------------------------------
# ‚ùå METHOD 5: PREPEND CHAR LOOP  üö´ WORST
# Time: O(n¬≤), Space: O(n)
# -----------------------------------------------------
rev5 = ""
for ch in s:
    rev5 = ch + rev5

print("Method 5 (WORST):", rev5)
```

```python
# =====================================================
# üîÅ LIST REVERSAL ‚Äî ALL METHODS (INTERVIEW READY)
# =====================================================

l = [1, 2, 3, 4, 5]

# -----------------------------------------------------
# ‚úÖ METHOD 1: reverse()  ‚≠ê BEST (In-place)
# Time: O(n), Space: O(1)
# -----------------------------------------------------
l1 = l.copy()
l1.reverse()
print("Method 1 (BEST - in-place):", l1)


# -----------------------------------------------------
# ‚úÖ METHOD 2: SLICING  ‚≠ê BEST (Clean & Pythonic)
# Time: O(n), Space: O(n)
# -----------------------------------------------------
l2 = l[::-1]
print("Method 2 (BEST - slicing):", l2)


# -----------------------------------------------------
# ‚úÖ METHOD 3: TWO-POINTER (DSA / FAANG LEVEL) ‚≠ê BEST
# Time: O(n), Space: O(1)
# -----------------------------------------------------
l3 = l.copy()
left, right = 0, len(l3) - 1

while left < right:
    l3[left], l3[right] = l3[right], l3[left]
    left += 1
    right -= 1

print("Method 3 (BEST - two pointer):", l3)


# -----------------------------------------------------
# ‚úÖ METHOD 4: reversed()  ‚≠ê GOOD
# Time: O(n), Space: O(n)
# -----------------------------------------------------
l4 = list(reversed(l))
print("Method 4 (GOOD):", l4)


# -----------------------------------------------------
# ‚ö†Ô∏è METHOD 5: INDEX LOOP + APPEND  ‚ö†Ô∏è AVERAGE
# Time: O(n), Space: O(n)
# -----------------------------------------------------
l5 = []
for i in range(len(l) - 1, -1, -1):
    l5.append(l[i])

print("Method 5 (AVERAGE):", l5)





```
