### 2 ways to get unique elements in *list*





```python
l=[1,2,3,1,2,4]
d={}
for a in l:
  if a in d:
    d[a]+=1
  else:
    d[a]=1
print(d)

keys_with_value_1 = [key for key, value in d.items() if value == 1]
print(keys_with_value_1)
```

```python
# @title add()
s = {1, 2, 3}
s.add(4)
print(s)  # {1, 2, 3, 4}
```

```python
# @title update()
s.update([5, 6])
print(s)  # {1, 2, 3, 4, 5, 6}
```

```python
# @title remove()
s.remove(3)
print(s)  # {1, 2, 4, 5, 6}
```

```python
# @title discard()
s.discard(10)  # No error if not present
print(s)
```

```python
# @title pop()
print(s)
element = s.pop()
print(element)  # Random element
print(s)
```

```python
# @title union()
a = {1, 2}
b = {2, 3}
print(a.union(b))  # {1, 2, 3}
c={9,10}
print(a|b|c)

```

```python
# @title intersection()
print(a.intersection(b))  # {2}
print(a & b)
```

```python
# @title difference() & difference_update()
print(a.difference(b))  # {1}
print(a-b)

a = {1, 2, 3}
b = {2, 3}
a.difference_update(b)
print(a)  # {1}
```

```python
# @title symmetric_difference() & symmetric_difference_update()
print(a.symmetric_difference(b))  # {1, 3}

a = {1, 2, 3}
b = {3, 4}
a.symmetric_difference_update(b)
print(a)  # {1, 2, 4}
```

```python
# @title issubset() , issuperset() , isdisjoint()
a = {1, 2}
b = {1, 2, 3}
print(a.issubset(b))  # True


print(b.issuperset(a))  # True


x = {1, 2}
y = {3, 4}
print(x.isdisjoint(y))  # True
```

### using loops

```python
def find_max_min_in_set(s):
  largest = float('-inf')
  smallest = float('inf')
  for i in s:
    if i >largest:
      largest=i
    if i<smallest:
      smallest=i
  return largest , smallest
s={9,0,1,6,2}
max_val, min_val = find_max_min_in_set(s)
print(f"max: {max_val}, min: {min_val}")
```
