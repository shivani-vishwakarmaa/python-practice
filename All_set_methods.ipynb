### 2 ways to get unique elements in *list*





```python
l=[1,2,3,1,2,4]
d={}
for a in l:
  if a in d:
    d[a]+=1
  else:
    d[a]=1
print(d)

keys_with_value_1 = [key for key, value in d.items() if value == 1]
print(keys_with_value_1)
```

```python
lst = [1, 2, 3, 1, 2, 4]
Counter(lst)

```

```python
from collections import Counter
lst=[1,2,3,1,2,4]
unique = list(Counter(lst).keys())
print(unique)
```

### removing duplicates in list

```python
l=[1,2,3,1,2,4]
print(set(l))
```

```python
l=[1,2,3,1,2,4]
remove_dup=[]
for i in l:
  if i not in remove_dup:
    remove_dup.append(i)
print( remove_dup)
```

```python
# @title add()
s = {1, 2, 3}
s.add(4)
print(s)  # {1, 2, 3, 4}
```

```python
# @title update()
s.update([5, 6])
print(s)  # {1, 2, 3, 4, 5, 6}
```

```python
# @title remove()
s.remove(3)
print(s)  # {1, 2, 4, 5, 6}
```

```python
# @title discard()
s.discard(10)  # No error if not present
print(s)
```

```python
# @title pop()
print(s)
element = s.pop()
print(element)  # Random element
print(s)
```

```python
# @title clear()
s.clear()
print(s)  # set()
```

```python
# @title union()
a = {1, 2}
b = {2, 3}
print(a.union(b))  # {1, 2, 3}
c={9,10}
print(a|b|c)

```

```python
# @title intersection()
print(a.intersection(b))  # {2}
print(a & b)
```

```python
# @title difference() & difference_update()
print(a.difference(b))  # {1}
print(a-b)

a = {1, 2, 3}
b = {2, 3}
a.difference_update(b)
print(a)  # {1}
```

```python
# @title symmetric_difference() & symmetric_difference_update()
print(a.symmetric_difference(b))  # {1, 3}

a = {1, 2, 3}
b = {3, 4}
a.symmetric_difference_update(b)
print(a)  # {1, 2, 4}
```

```python
# @title issubset() , issuperset() , isdisjoint()
a = {1, 2}
b = {1, 2, 3}
print(a.issubset(b))  # True


print(b.issuperset(a))  # True


x = {1, 2}
y = {3, 4}
print(x.isdisjoint(y))  # True
```

```python
# @title copy()
s = {5, 6, 7}
s_copy = s.copy()
print(s_copy)  # {5, 6, 7}
```

```python

# @title len() on set
print(len(s))  # 3
```

```python
# @title in keyword
print(5 in s)   # True
print(9 in s)   # False
```
